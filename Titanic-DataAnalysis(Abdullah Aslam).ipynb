```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
color_pallete = [ "#00876c",
                "#56a474",
                "#8ebf7f",
                "#c6da8e",
                "#fff4a4",
                "#f9ca7a",
                "#f29e5e",
                "#e77051",
                "#d43d51"]
sns.set_style('whitegrid')
sns.set_palette(color_pallete)
```

```python
pathToData = './Data/Titanic.csv'
titanicData = pd.read_csv(pathToData)
```

```python
titanicData.describe()
```

```python
titanicData.head()
```

```python
titanicData.tail()
```

# Survived passengers

```python
plt.figure(figsize=(14, 6))

plt.title("Distribution of survived passengers")
plt.ylabel('People count')

sns.countplot(titanicData['Survived'])

plt.show()
```

# Gender of people 

```python
plt.figure(figsize=(14, 6))

plt.title("Distribution of gender of passengers")
plt.xlabel('Gender')
plt.ylabel('People count')

sns.countplot(titanicData['Sex'])

plt.show()
```

# Siblings or Spouse

```python
plt.figure(figsize=(14, 6))

sns.countplot(titanicData['SibSp'])

plt.title("Distribution of Siblings/Spouse")
plt.xlabel('No of siblings or spouse')
plt.ylabel('People count')

plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.title("Distribution of Embarked Location")
plt.xlabel('Location')
plt.ylabel('People count')

sns.countplot(titanicData['Embarked'])

plt.show()
```

```python
plt.figure(figsize=(14, 6))

sns.barplot(x=titanicData['SibSp'], y=titanicData['Survived'])

plt.title("Survival based on Siblings/Spouse")
plt.xlabel("Siblings/Spouse")
plt.ylabel("Chance of survival")

plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.title("Distribution of age")
sns.kdeplot(titanicData['Age'], shade=True)


plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.title("Distribution of fare")

sns.distplot(titanicData['Fare'])

plt.show()
```

```python
plt.figure(figsize=(14, 6))

sns.catplot('Pclass',data=titanicData,hue='Sex',kind='count')

plt.title("Genderwise passenger class")
plt.xlabel("Passengers' class")
plt.ylabel("People count")

plt.show()
```

```python
plt.figure(figsize=(14, 6))

sns.catplot('Embarked',data=titanicData,hue='Pclass',kind='count')

plt.title("Passenger class wise embarkment")
plt.xlabel("Passengers' classes")plt.ylabel("People count")

plt.show()
```

```python
plt.figure(figsize=(14, 6))

sns.catplot('Embarked',data=titanicData,hue='Survived',kind='count')

plt.title("Survival based on Embarkment")
plt.xlabel("Location of Embarkment")
plt.ylabel("People count")

plt.show()
```

```python
plt.figure(figsize=(14, 6))

sns.catplot('SibSp', data=titanicData, hue='Survived', kind='count')

plt.title("Survival based on no of Siblings/Spouse")
plt.xlabel("Siblings/Spouse")
plt.ylabel("People count")

plt.show()
```

```python
plt.figure(figsize=(14, 6))

sns.catplot('Embarked', data=titanicData, hue='Sex', kind='count')

plt.title("Embarment based on Gender")
plt.xlabel("Location of Embarkment")
plt.ylabel("People count")

plt.show()
```

```python
plt.figure(figsize=(14, 6))

sns.scatterplot(x=titanicData['Fare'], y=titanicData['Age'])

plt.show()
```

```python
plt.figure(figsize=(14, 6))

sns.lmplot(x='Age', y='Pclass', data=titanicData)

plt.title("Age based on passenger class")
plt.ylabel("Passenger Class")

plt.show()
```
